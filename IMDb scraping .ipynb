```python

# Define the base URL and number of pages to scrape,understanding that we need multiple URL's to swap 
base_url = "https://www.imdb.com/search/title/?groups=top_1000&sort=user_rating,desc&count=100&start="
num_pages = 5  # Changing this to the desired number of pages required to do the scraping
```

```python
# Defining empty lists to store data
movie_name = []
year = []
runtimes = []  
rating = []
metascore = []
votes = []
gross = []
description = []


# Looping through the pages from the current page to the 5 page, since the current page doesnt have a start value,
#and since we are starting with one doing page-1 helps in maintaining the loop

for page in range(1, num_pages + 1):
    # Constructing the basic URL structure for the current page,num_pages determines the number of pages to scrape, 
    #allowing for dynamic navigation.
    
    url = f"{base_url}{(page - 1) * 100 + 1}"

    # Send an HTTP GET request to the URL, to determine if the URL is returning a successful return statement
    response = requests.get(url)

    # Check if the request was successful
    if response.status_code == 200:
        # Parse the HTML content of the page
        soup = BeautifulSoup(response.text, 'html.parser')

# Extracting the data we need from the current page and append it to the lists while exploring the page all of the movie details had come under a single class tag called as. "lister-item-content"
#Therefore a single for loop was sufficient to append it into a list and make a data frame out of it
        
        
        
    for m in soup.find_all("div", class_="lister-item-content"):
       
        #finding out and taking all the title names and appending in the list
        title = m.h3.a.text
        movie_name.append(title)
            
        #years in the IMDb database and appending them to the new list
        year_elem = m.h3.find("span", class_="lister-item-year").text.strip('()')
        year.append(year_elem)

        #rating
        imdb_rating = m.strong.text
        rating.append(imdb_rating)
            
        #found the runtime class in the p tag of the html file
        runtime = m.p.find("span", class_="runtime").text if m.p.find("span", class_="runtime") else None
        runtimes.append(runtime)
        
        #found the metascore class in the span element of the html file
        metascore_elem = m.find("span", class_="metascore").text.strip() if m.find("span", class_="metascore") else None
        metascore.append(metascore_elem)
            
     # Extract the number of votes for the movie from a <span> element with the name attribute set to "nv"
        num_votes = m.find("span", {"name": "nv"})["data-value"]
        votes.append(num_votes)

    # Extract the gross box office earnings of the movie (if available)
    # The second <span> element with the name attribute set to "nv" 
    # Checking if there are at least two such elements, and if so, extract the second one 
        gross_elem = m.find_all("span", {"name": "nv"})[1]["data-value"] if len(m.find_all("span", {"name": "nv"})) > 1 else None
        gross.append(gross_elem)

    # Extract the plot description of the movie
    # The last <p> element with the class "text-muted" contains the plot summary
    # If such elements are found, extract the text and remove leading/trailing whitespace; otherwise, assign None
        plot = m.find_all("p", class_="text-muted")[-1].text.strip() if m.find_all("p", class_="text-muted") else None
        description.append(plot)

                                      

    # Creating a DataFrame from the lists
    data = pd.DataFrame({"Movie Name": movie_name,"Year of release": year,"Time": runtimes, "Rating": rating,
    "Metascore": metascore, "Votes": votes,"Gross Margin": gross,"Description": description
    })
data
```

```python
# Specifing the file path where you want to save the CSV file
csv_file_path = 'movie_data.csv'

# Exporting the DataFrame to a CSV file
data.to_csv(csv_file_path, index=False)

print(f'DataFrame saved to {csv_file_path}')
```
